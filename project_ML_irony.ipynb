```python
from sklearn import preprocessing
from sklearn.model_selection import cross_val_score
```

```python
f = open("C:/Users/1/Desktop/ирония.csv", "r", encoding = "Windows-1251")
lines = f.readlines()
```

```python
dic = {}
all_words = []
for line in lines:
    line = line.split(";")
    words = line[1].split()
    for word in words:
        word = word.strip(",")
        word = word.strip(":")
        word = word.strip("!")
        word = word.strip("?")
        word = word.strip("...")
        word = word.strip(".")
        if word not in dic:
            dic[word] = 1
        else:
            dic[word] += 1
for key, value in dic.items():
    if value < 20 and value > 1:
        all_words.append(key)
print (all_words)
```

```python
lines_new = []
for line in lines:
    if line != lines[550]:
        lines_new.append(line)
matrix = []
vector = []
for line in lines_new:
    arr = []
    line = line.split(";")
    if line[0][0] == '0' or line[0][0] == '1':
        vector.append(int(line[0]))
    twit = line[1]
    words_twit = twit.split()
    for word_new in all_words:
        count = 0
        for word_twit in words_twit:
            word_twit = word_twit.strip(",")
            word_twit = word_twit.strip(":")
            word_twit = word_twit.strip("!")
            word_twit = word_twit.strip("?")
            word_twit = word_twit.strip("...")
            word_twit = word_twit.strip(".")
            if word_new == word_twit:
                count += 1
        arr.append(count)
    matrix.append(arr)
```

```python
matrix = preprocessing.scale(matrix)
print (matrix)
```

```python
matrix_train = matrix[0:400]
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
```

```python
clf = svm.SVC()
clf.fit(matrix_train, vector)
```

```python
predictions = clf.predict(matrix[400:])
print (predictions)
```

```python
scores = cross_val_score(clf, matrix[:400], vector, cv=5)
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```
